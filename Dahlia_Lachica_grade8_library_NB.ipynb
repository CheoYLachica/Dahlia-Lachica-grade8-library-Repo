```python
import firebase_admin
from firebase_admin import credentials, db
cred = credentials.Certificate("/content/firebase_key.json.json")
firebase_admin.initialize_app(cred, {
 "databaseURL": "https://dlia-lachica-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"
})
print("Firebase connected successfully!")
```

```python
import json
with open("/content/book.json", "r") as file:
 data = json.load(file)
print("JSON file loaded")
ref = db.reference("books") # Changed from "students" to "books"
for book_item in data["books"]: # Changed from data["students"] to data["books"]
 ref.child(str(book_item["id"])).set(book_item) # Using book_item["id"] assuming books have an 'id'
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
for book_id, book_info in books_data.items():
 print(book_info["title"])
```

```python
for book_id, book_info in books_data.items():
 print(book_info["title"], book_info["details"]["available"])
```

```python
total_books = len(books_data)
print(f"Total number of books in the library: {total_books}")
```

```python
total_copies = 0
for book_id, book_info in books_data.items():
    total_copies += book_info['details']['copies']

print(f"Total number of copies of all books in the library: {total_copies}")
```
